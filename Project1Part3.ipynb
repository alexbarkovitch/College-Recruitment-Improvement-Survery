## Question 2: Compare the gender demographics of the different classes who answered the survey. Are there any significant differences?

```python
#see how many people of each gender took this survery
gender = df.gender.value_counts()
gender
#from this data, you can see that mostly women took this survey
```

```python
#groups genders for each course
gender_summary = df.groupby('course')['gender'].value_counts().reset_index()
gender_summary
```

```python
#plot more specific gender information by class on the x-axis
#make a stacked bar graph of how many people of each gender are in each course
#start by aggregating the data by course and gender
course_gender_counts = df.groupby(['course', 'gender']).size().unstack(fill_value=0)

#plot the stacked bar graph with adjusted bar width
course_gender_counts.plot(kind='bar', stacked=True, width=0.8)

#title and label the graph
plt.title('Number of Students by Gender for Each Course')
plt.xlabel('Course')
plt.ylabel('Number of Students')

#rotate x-axis labels so you can read them clearly
plt.xticks(rotation=45, ha='right') 
#source: https://stackoverflow.com/questions/10998621/rotate-axis-tick-labels

#place the legend outside the plot area
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
#source: https://stackoverflow.com/questions/4700614/how-to-put-the-legend-outside-the-plot

#show graph
plt.show()
```

```python
#another way to look at the above graph, sorts by gender on the x-axis
#aggregate the data by course and gender
course_gender_counts = df.groupby(['gender', 'course']).size().unstack(fill_value=0)

#plotting the stacked bar graph with adjusted bar width
course_gender_counts.plot(kind='bar', width=0.8)  #makes the bar wider

#title and label the graph
plt.title('Number of Students by Course for Each Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')

#rotate x-axis labels so you can read them clearly
plt.xticks(rotation=45, ha='right') 
#source: https://stackoverflow.com/questions/10998621/rotate-axis-tick-labels

#place the legend outside the plot area
plt.legend(title='Course', bbox_to_anchor=(1.05, 1), loc='upper left')
#source: https://stackoverflow.com/questions/4700614/how-to-put-the-legend-outside-the-plot

#display graph
plt.show()
```

## Question 3: What are the most and least common ways people heard about CCM?

### For this question, we want to look at how many yes/no responses each column got.

```python
#filter out the columns which begin with 'heard'
heard_columns = df.filter(like='heard').columns
heard_columns
```

```python
#add up amoount of 'yes' responses for each column
heard_counts = df[heard_columns].apply(lambda x: (x=='Yes').sum())
heard_counts
```

```python
#sort and display the data in a dataframe
heard_simple_view = pd.DataFrame({'Heard about CCM': heard_counts.index, 'Yes Count': heard_counts.values})
#make the dataframe in order of most yes responses to least and tanspose so it's horizontal
heard_simple_view_sorted = heard_simple_view.sort_values(by = 'Yes Count', ascending = False).transpose()
heard_simple_view_sorted
```

```python
#to graph the data, we'll count up all yes/no/don't recall response for each column
#transpose so that heard methods are on the x-axis of the graph
heard_responses = df[heard_columns].apply(pd.Series.value_counts).transpose()
heard_responses
```

```python
#create a stacked bar chart to visualize the response students gave for how they heard about ccm
heard_responses.plot(kind='barh', stacked=True)

#title and label the chart
plt.title('Responses for How Students Heard About CCM')
plt.xlabel('Heard Method')
plt.ylabel('Number of Responses')

#rotate x-axis labels
plt.xticks(rotation=45, ha='right')
#source: https://stackoverflow.com/questions/10998621/rotate-axis-tick-labels

#place the legend outside the plot area
plt.legend(title='Resonse Type', bbox_to_anchor=(1.05, 1), loc='upper left')
#source: https://stackoverflow.com/questions/4700614/how-to-put-the-legend-outside-the-plot

#makes the bars tighter
plt.tight_layout()
#source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html

#view graph
plt.show()
```

#### This data shows us that the top 3 ways that people heard about CCM are through friends and family (75), high school counselor (53), and through a current student (47). The least common ways people heard about CCM are through the radio (5), employer (5), and from television (10).
#### The gender demographics also show us some interesting things. While for most of the options more women gave yes responses than men, it makes sense considering there were more women who took the survery. However, you can see that significantly more men heard about CCM through an app advertisement and from the radio than women. Men and women heard about CCM about equally through websites, and many more women heard about CCM from friends/family and from high school counselors.
#### This information could allow CCM to do more targeted advertising, and put more or less investment into different advertising methods based on this data.

## Question 4: What are the most and least common motivations for people to seek a computing class at CCM?

### This question also looks at the amount of yes/no responses for each question.

```python
#now we want to look at what motivated students to attend computer courses at CCM
#group all the columns which start with 'motivation'
motivation_columns = df.filter(like='motivation').columns
motivation_columns
```

```python
#count up how many 'yes' responses are under each column
motivation_counts = df[motivation_columns].apply(lambda x: (x=='Yes').sum())
motivation_counts
```

```python
#sort and display the data in a dataframe
motivation_summary = pd.DataFrame({'Motivation to take CCM courses': motivation_counts.index, 'Yes Count': motivation_counts.values})
#make the dataframe in order of most yes responses to least and transpose so it's horizontal
motivation_sorted = motivation_summary.sort_values(by = 'Yes Count', ascending = False).transpose()
motivation_sorted
```

```python
#to graph the data, count each yes/no response for each column
#transpose again so the motivations are on the x-axis
motivation_responses = df[motivation_columns].apply(pd.Series.value_counts).transpose()
motivation_responses
```

```python
#create a stacked bar chart to visualize the response students gave for what motivated them to pick a computing class at ccm
motivation_responses.plot(kind='barh', stacked=True)

#title and label the graph
plt.title('Responses for What Motivated Students to Take a Computing Class at CCM')
plt.xlabel('Motivation')
plt.ylabel('Number of Responses')

#rotate x-axis labels
plt.xticks(rotation=45, ha='right')
#source: https://stackoverflow.com/questions/10998621/rotate-axis-tick-labels

#place the legend outside the plot area
plt.legend(title='Resonse Type', bbox_to_anchor=(1.05, 1), loc='upper left')
#source: https://stackoverflow.com/questions/4700614/how-to-put-the-legend-outside-the-plot

#make the bars tighter
plt.tight_layout()
#source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html

#view graph
plt.show()
```

#### This data shows us that by far the greatest motivation for students seeking a computing course at CCM is that it's a required class (90). This makes sense because the survey is based on non-major students, whose interests more likely lie in their own majors. For students majoring in other subjects, there were still many with motivations aside from the requirement. There were many people who registered for a computing class because they wanted to keep current with computer skills (55), and to help with career advancement (44). 
#### The certifications offered (7) and relocation (4) were the least chosen motivations. Students that aren't computer science majors don't have much need for a computer science certificate, and community college isn't something that many would relocate for.

```python
interest_2020 = df_2020.groupby('interest_level_2020')['course'].count().reset_index()
interest_2020
```

```python
#create a pie chart
plt.pie(interest_summary['course'], labels = ['1','2','3','4','5'])

#title the chart
plt.title("Level of Interest in Taking More Courses, With 1 Being the Least and 5 the Most Interested")
```

```python
#create a bar chart to visualize interest level data from 2023
plt.bar(interest_summary.interest_level_2023, interest_summary.course, color=['c','g','r','b', 'm'])

#title and label the graph
plt.title("Level of Interest in Taking More Courses, With 1 Being the Least and 5 the Most Interested from 2023 Survey")
plt.xlabel("Interest Level")
plt.ylabel("Number of Students")
```

```python
#create a bar chart to visualize interest level data from 2020
plt.bar(interest_2020.interest_level_2020, interest_2020.course, color=['c','g','r','m', 'b'])

#title and label the graph
plt.title("Level of Interest in Taking More Courses, With 1 Being the Least and 5 the Most Interested from 2020 Survey")
plt.xlabel("Interest Level")
plt.ylabel("Number of Students")
```

```python
#combine the two plots into a subplot
fig, axs = plt.subplots(2)
#title the plot
fig.suptitle('Interest Level from Different Years')
#plot and label the top graph
axs[0].plot(interest_summary.interest_level_2023, interest_summary.course)
axs[0].set_title("2023 Survey")
#plot and label the bottom graph
axs[1].plot(interest_2020.interest_level_2020, interest_2020.course)
axs[1].set_title("2020 Survey")
```

#### This data shows us that the interest level for non-majors to take more computer science courses is extremely low. In 2023, most people were neutral, but there were many more people who answered that they aren't interested than those who are interested in taking more classes. Considering the earlier analysis about motivations, non-major students probably only need one or two courses to get what they need, and so they aren't interested in taking more. By comparing the interest levels of students in 2020 and in 2023, the results are very similar. In 2020, 3 was still the most picked interest level, but 2 was only behind by one person. 1 and 2 had many more picks than 4 and 5. Overall, the majority of students are not interested in taking more computing classes.  You can also see this trend in the fact that there were many more people who took the survery in 2020 than in 2023. Maybe taking the results of the analysis on these surveys will allow CCM to raise the interest level of non-majors to encourage more people to register for computing classes.
